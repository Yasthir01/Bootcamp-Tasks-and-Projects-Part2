```python
import pandas as pd
```

```python
df = pd.read_table('balance.txt', delim_whitespace=True)
```

```python
df
```

```python
# average income based on ethnicity
df[['Income', 'Ethnicity']].mean()
```

```python
# on average, do married or single people have a higher balance
df.groupby(['Married']).mean().sort_values('Balance', ascending=False)
```

```python
# highest income
df['Income'].max()
```

```python
# lowest income
df['Income'].min()
```

```python
# number of cards in dataset
df['Cards'].sum()
```

```python
# How many females do we have information for vs how many males
df.groupby(['Gender']).count()
```
